```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 엑셀 파일 경로 지정
file_path = '/content/drive/MyDrive/Colab Notebooks/boston_house_data.xlsx'

# pandas를 사용하여 데이터 읽어오기
data = pd.read_excel(file_path)

# 12번 열 데이터를 특징으로 선택 (인덱스는 0부터 시작)
x = data.iloc[:, 11].values.reshape(-1, 1)
y = data.iloc[:, 11].values.reshape(-1, 1)

# 데이터를 학습 데이터와 테스트 데이터로 분할 (7:3 비율)
train_size = int(len(x) * 0.7)
x_train, x_test = x[:train_size], x[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# 학습 데이터에 bias term 추가
A = np.hstack([np.ones(x_train.shape[0]).reshape(-1, 1), x_train])
A = np.asmatrix(A)

# 가중치 초기화 (w1: 기울기, w0: y절편)
w1 = 0
w0 = 0

# 학습률 및 에포크 설정
learning_rate = 0.01
epochs = 10000

# 입력 데이터 개수
n = float(len(x_train))

# 경사 하강법
for i in range(epochs):
    y_pred = w1 * x_train + w0  # 예측값 계산
    dw1 = (1/n) * np.sum(x_train * (y_pred - y_train))  # 기울기 업데이트 (np.sum 사용)
    dw0 = (1/n) * np.sum(y_pred - y_train)  # y절편 업데이트 (np.sum 사용)
    w1 = w1 - learning_rate * dw1
    w0 = w0 - learning_rate * dw0

# 학습된 모델 시각화
plt.figure(figsize=(6, 4))
plt.title('Linear Regression (Gradient Descent)', fontsize=16)
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.scatter(x_train, y_train, color='blue', label='Training data')  # 학습 데이터 시각화
x_line = np.linspace(0, 40, 100).reshape(-1, 1)  # x 범위를 0~40으로 설정, reshape 추가
y_line = w1 * x_line + w0
plt.plot(x_line, y_line, color='red', label='Regression line')  # 회귀선 시각화

# x축과 y축 범위 설정
plt.xlim(0, 40)
plt.ylim(0, 50)

plt.legend()
plt.grid(alpha=0.4)
# plt.ion()  # interactive mode 활성화 (선택 사항)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# 엑셀 파일 경로 지정
file_path = '/content/drive/MyDrive/Colab Notebooks/boston_house_data.xlsx'

# pandas를 사용하여 데이터 읽어오기
data = pd.read_excel(file_path)

# 12번 열 데이터를 특징으로 선택 (인덱스는 0부터 시작)
X = data.iloc[:, 12].values.reshape(-1, 1)
y = data.iloc[:, 0].values

# 데이터를 학습 데이터와 테스트 데이터로 분할 (7:3 비율)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 학습 데이터에 bias term 추가
X_train = np.hstack([np.ones(X_train.shape[0]).reshape(-1, 1), X_train])

# 가중치 초기화 (w1: 기울기, w0: y절편)
w1 = 0
w0 = 0

# 학습률 및 에포크 설정
learning_rate = 0.01
epochs = 10000

# 입력 데이터 개수
n = float(len(X_train))

# 경사 하강법
for i in range(epochs):
    y_pred = w1 * X_train[:, 1] + w0  # 예측값 계산
    dw1 = (1/n) * sum(X_train[:, 1] * (y_pred - y_train))  # 기울기 업데이트
    dw0 = (1/n) * sum(y_pred - y_train)  # y절편 업데이트
    w1 = w1 - learning_rate * dw1
    w0 = w0 - learning_rate * dw0

# 학습된 모델 시각화
plt.figure(figsize=(6, 4))
plt.title('Linear Regression (Gradient Descent)', fontsize=16)
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.scatter(X_train[:, 1], y_train, color='blue', label='Training data')  # 학습 데이터 시각화
x_line = np.linspace(X_train[:, 1].min(), X_train[:, 1].max(), 100)
y_line = w1 * x_line + w0
plt.plot(x_line, y_line, color='red', label='Regression line')  # 회귀선 시각화
plt.legend()
plt.grid(alpha=0.4)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# 데이터 불러오기
data = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/boston_house_data.xlsx')
x = data.iloc[:, 11].values.reshape(-1, 1)  # 12번 열 데이터
y = data.iloc[:, 0].values # 목표 변수 (필요에 따라 변경)


# 경사 하강법 함수 (2, 3, 4차 모델에 사용)
def gradient_descent(x, y, degree, learning_rate=0.001, epochs=10000):
    X = np.c_[np.ones(x.shape[0]), x]
    for i in range(2, degree + 1):
        X = np.c_[X, x**i]  # 다항식 특성 추가
    w = np.zeros(X.shape[1])
    for i in range(epochs):
        y_pred = X @ w
        error = y_pred - y
        gradient = (1/len(x)) * X.T @ error
        w = w - learning_rate * gradient
    return w, X


# 2차 모델 (경사 하강법)
w, X = gradient_descent(x, y, degree=2)

# x_line 범위를 데이터 x 범위와 일치시키고 reshape
x_line = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)

# X_line 생성 시 bias term과 다항식 항 추가
X_line = np.c_[np.ones(x_line.shape[0]), x_line, x_line**2]

y_line = X_line @ w  # 회귀선 계산
plt.scatter(x, y, label='Data')
plt.plot(x_line, y_line, color='red', label='Regression line')
plt.title('2차 모델 (경사 하강법)')
plt.legend()
plt.show()
```

```python
!pip install -U scikit-learn
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# CSV 파일에서 데이터 로드
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/multivariate_data.csv')

# Class 1 and 4 데이터 추출
class1_data = data[['C1_X1', 'C1_X2']]
class4_data = data[['C4_X1', 'C4_X2']]
class1_data['class'] = 1  # Class 1에 대한 target variable 설정
class4_data['class'] = 0  # Class 4에 대한 target variable 설정
class1_data.columns = ['X1', 'X2', 'class']  # 열 이름 변경
class4_data.columns = ['X1', 'X2', 'class']  # 열 이름 변경

# 데이터 합치기
selected_data = pd.concat([class1_data, class4_data])

# 특징(X)과 target variable(y) 분리
X = selected_data[['X1', 'X2']]
y = selected_data['class']

# 학습 데이터와 테스트 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 로지스틱 회귀 모델 생성 및 학습
model = LogisticRegression(max_iter=10000)  # 에포크 설정
model.fit(X_train, y_train)

# 시각화
plt.figure(figsize=(8, 6))
plt.scatter(X_train[y_train == 1]['X1'], X_train[y_train == 1]['X2'], color='red', label='Class 1')
plt.scatter(X_train[y_train == 0]['X1'], X_train[y_train == 0]['X2'], color='blue', label='Class 0')

# Decision boundary 시각화
x_min, x_max = X_train['X1'].min() - 1, X_train['X1'].max() + 1
y_min, y_max = X_train['X2'].min() - 1, X_train['X2'].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.3)  # Decision boundary를 색상으로 표시

plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Logistic Regression for Class 1 and 4')
plt.legend()
plt.grid(alpha=0.4)
plt.show()
```
