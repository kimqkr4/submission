```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

A = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data0405.csv')
x = A['x'].values
y = A['y'].values

X = np.vstack([np.ones_like(x), x]).T

w = np.linalg.solve(X.T @ X, X.T @ y)

x_plot = np.linspace(min(x), max(x), 200).reshape(-1, 1)
X_plot = np.hstack([np.ones_like(x_plot), x_plot])
y_pred = X_plot @ w

plt.scatter(x, y, color='black', label='data')
plt.plot(x_plot, y_pred, color='blue', label='LS deg 1')
plt.title('Least Squares Degree 1')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()

print('Parameters w (1차 모델):')
print(w)
```

```python
X = np.vstack([x**0, x, x**2]).T

w = np.linalg.solve(X.T @ X, X.T @ y)

x_plot = np.linspace(min(x), max(x), 200).reshape(-1, 1)
X_plot = np.hstack([x_plot**0, x_plot, x_plot**2])
y_pred = X_plot @ w

plt.scatter(x, y, color='black', label='data')
plt.plot(x_plot, y_pred, color='blue', label='LS deg 2')
plt.title('Least Squares Degree 2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()

print('Parameters w (2차 모델):')
print(w)
```

```python
X = np.vstack([x**i for i in range(11)]).T

w = np.linalg.solve(X.T @ X, X.T @ y)

x_plot = np.linspace(min(x), max(x), 200).reshape(-1, 1)
X_plot = np.hstack([x_plot**i for i in range(11)])
y_pred = X_plot @ w

plt.scatter(x, y, color='black', label='data')
plt.plot(x_plot, y_pred, color='blue', label='LS deg 10')
plt.title('Least Squares Degree 10')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()

print('Parameters w (10차 모델):')
print(w)
```

```python
X = np.vstack([x**0, x]).T

w = np.zeros(2)

learning_rate = 0.001
iterations = 500

for i in range(iterations):
    y_pred = X @ w
    error = y_pred - y
    gradient = X.T @ error / len(x)
    w = w - learning_rate * gradient

x_plot = np.linspace(min(x), max(x), 200).reshape(-1, 1)
X_plot = np.hstack([x_plot**0, x_plot])
y_pred = X_plot @ w

plt.scatter(x, y, color='black', label='data')
plt.plot(x_plot, y_pred, color='blue', label='GD deg 1')
plt.title('Gradient Descent Degree 1')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()

print('Parameters w (1차 모델):')
print(w)
```
